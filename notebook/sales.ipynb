Data Cleaning

```python
#Count Missing Values
df.isnull().sum()
```

```python
# Fill Missing Values
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
```

```python
df["City"] = df["City"].fillna("Unknown")
```

```python
#Remove Duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# convert date
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
# Check Unique Values
df["Category"].unique()
df["Category"].value_counts()
```

Feature Engineering

```python
df["Year"] = df["Order_Date"].dt.year
df["Month"] = df["Order_Date"].dt.month_name()
df["Day"] = df["Order_Date"].dt.day_name()
```

```python
#Profit Percentage
df["Profit %"] = (df["Profit"] / df["Sales"]) * 100
```

```python
#Total Sales
df["Profit"].sum()
```

```python
#Average Sales
df["Sales"].mean()
```

```python
#Highest Sales
df["Sales"].max()
```

```python
#Lowest Sales
df["Sales"].min()
```

```python
#Total Orders
df["Order_ID"].count()
```

```python
# Top Categories
df.groupby("Category")["Sales"].sum()
```

```python
#Top Cities
df.groupby("City")["Sales"].sum().sort_values(ascending=False)
```

```python
#Month Sales
df.groupby("Month")["Sales"].sum()
```

```python
#Top 10 Prroducts
df.groupby("Product")["Sales"].sum().nlargest(10)
```

Data Visualiation

```python
#Sales by Category
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

```python
#Monthly Sales
df.groupby("Month")["Sales"].sum().plot(kind="line")
plt.title("Monthly Sales")
plt.show()
```

```python
#Category Distribution
df["Category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python
#Sales vs Profit
plt.scatter(df["Sales"], df["Profit"])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
#Sales
plt.boxplot(df["Sales"])
plt.show()
```
